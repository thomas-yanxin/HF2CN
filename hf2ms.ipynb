```python
# 您准备从Hugging Face下载的模型的名称
hf_repo_id = "X-D-Lab/MindChat-Qwen-7B-v2"
# 您准备传入ModelScope的模型的名称
ms_model_id = "thomas/MindChat-Qwen-7B-v2"
```

```python
# 登录Hugging Face 
# https://huggingface.co/settings/tokens
from huggingface_hub import login
login()
```

```python
from huggingface_hub import snapshot_download
import os
# 从HuggingFace下载模型
local_dir = os.path.join(os.path.abspath('.'), hf_repo_id.split("/")[1])
snapshot_download(repo_id=hf_repo_id, local_dir=local_dir)
```

```python
!touch configuration.json
!mv configuration.json {local_dir}
```

```python
# 登录ModelScope
from modelscope.hub.api import HubApi

# 用您的账号登录https://www.modelscope.cn ，在个人中心->访问令牌, 复制SDK访问令牌
YOUR_ACCESS_TOKEN = 'YOUR_MODELSCOPE_SDK_ACCESS_TOKEN'

api = HubApi()
api.login(YOUR_ACCESS_TOKEN)
```

```python
# 将模型上传到ModelScope
api.push_model(
    model_id=ms_model_id,
    model_dir=local_dir, # 本地模型目录，要求目录中必须包含configuration.json
    visibility=1 # 1: private, 5: public
)
```
